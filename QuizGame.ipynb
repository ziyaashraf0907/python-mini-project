```python

# Simple Quiz Game in Python using for loop, match-case, and if-else

def askQuestion(question, options, correct_answer):
    print(question)
    for i in range(len(options)):
        print(f'{i+1}.{options[i]}')


    answer = input("Choose the correct option (1/2/3/4): ")

# Match-case for handling answer selection
    match answer:
        case "1":
            selected_answer = options[0]
        case "2":
            selected_answer = options[1]
        case "3":
            selected_answer = options[2]
        case "4":
            selected_answer = options[3]
        case _:
            print("Invalid input! Please choose a number between 1 and 4.\n")
            return False  # Return False if invalid input

# Check answer using if-else
    if selected_answer == correct_answer:
        print(f"Correct ! You got 10 points\n")
        return True
    else:
        print(f"Incorrect ! You got 0 points\n The correct answer is: {correct_answer}\n ")
        return False

def quizGame():
    questions = [
        {
            "question": "What is the basic data type for numbers in python?",
            "options": ["int", "str", "char", "bool"],
            "correct_answer": "int"
        },
        {
            "question": "Who developed python programming language?",
            "options": ["Wick van Rossum", "Rasmus Lerdorf", "Guido van Rossum", "Niene Stom"],
            "correct_answer": "Guido van Rossum"
        },
        {
            "question": "What is the basic control structure in python?",
            "options": ["for", "while", "switch", "if"],
            "correct_answer": "if"
        },
        {
            "question": "Which is the following correct extension of python?",
            "options": [".python", ".pl", ".py", ".p"],
            "correct_answer": ".py"
         },
         {
            "question": "Which keyword is used for function in python language?",
            "options": ["Function", "def", "Fun", "Define"],
            "correct_answer": "def"
          },
          {
            "question": "What is the basic data structure for lists in python?",
            "options": ["tuple", "list", "dictionary", "set"],
            "correct_answer": "list"
          },
          {
            "question": "What is the function to get user input in python?",
            "options": ["input()", "print()", "raw_input()", "get_input()"],
            "correct_answer": "input()"
          },
          {
            "question": "What is the print function in python?",
            "options": ["print[]", "print()", "print{}", "print_()"],
            "correct_answer": "print()"
          },
          {
            "question": "What is basic data type for text in python?",
            "options": ["int", "float", "string", "list"],
            "correct_answer": "string"
          },
          {
            "question": "What is the basic syntax to comment in python?",
            "options": ["//", "#", "/*", "*/"],
            "correct_answer": "#"
          }

    ]

    score = 0

    for q in questions:
        if askQuestion(q["question"], q["options"], q["correct_answer"]):
            score += 10

    print(f" Your final score is {score}/100.")

if __name__ == "__main__":
    quizGame()
```
