```python
#request- library in python-method/functions
#That enables user to PERFORM CRUD operations
#with any server or database
import requests
```

```python
url = 'https://jsonplaceholder.typicode.com/posts'
```

```python
response = requests.get(url=url)
```

```python
response.json()
```

```python
type(response.json())
```

```python
for _ in response.json():
   print(f'userId is {_["userId"]}')
```

```python
for _ in response.json():
    print(f'post number is {_["id"]}')
```
