```python
ScoredByBrazil = df['home_team'] =='Brazil'
```

```python
df[ScoreAboveZero & ScoredByBrazil]
```

```python
brazildf = df[ScoreAboveZero & ScoredByBrazil].sort_values(by='home_score',ascending=True)
```

```python
brazildf
```

```python
brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x='home_score',y='away_score',kind = 'scatter')
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.legend('Takeaway')
plt.title('Brazil Matches')
plt.show()
```

```python
brazildf.plot(x='Year',y='home_score',kind = 'hexbin')
plt.xlabel('Year')
plt.ylabel('home score')
plt.legend('Takeaway')
plt.title('Brazil Matches')
plt.show()
```
